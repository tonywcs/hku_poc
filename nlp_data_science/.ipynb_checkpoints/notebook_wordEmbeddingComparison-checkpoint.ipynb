```python
from sentence_transformers import SentenceTransformer, util
from sklearn.datasets import fetch_20newsgroups
import pandas as pd
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from matplotlib import pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn import metrics
from sklearn.metrics.cluster import contingency_matrix
import spacy
import spacy_universal_sentence_encoder

import numpy as np
from joblib import Parallel, delayed
from sklearn.utils import check_random_state
from sklearn.metrics.pairwise import distance_metrics
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.metrics.pairwise import cosine_similarity
from joblib import Parallel, delayed
from itertools import combinations

USE_model = spacy_universal_sentence_encoder.load_model('en_use_lg')
Word2Vec_model = spacy.load('en_core_web_lg')
BERT_model = SentenceTransformer('paraphrase-distilroberta-base-v1')
```

```python
# Dataset

# 20 news groups
def get20newsgroups():
    newsgroups = fetch_20newsgroups()
    df = pd.DataFrame(data = {'news': newsgroups.data, 'tag': newsgroups.target})
    df = df.groupby('tag', group_keys=False).apply(lambda x: x.sample(min(len(x), 50)))
    display(df.groupby('tag')['news'].nunique())
    return list(df['news']), list(df['tag'])

# custom dataset
def getCSV():
    df = pd.read_csv('custom_dataset.csv', encoding = "mac-roman")
    return list(df['Title']), list(df['Tag'])
```

```python
# Word Embeddings Models

# SentenceBERT
def getSentenceBERT(text):
    embeddings = BERT_model.encode(text, show_progress_bar=True, convert_to_numpy=True)
    return embeddings

# TFIDF
def getTFIDF(text):
    vectorizer = TfidfVectorizer(stop_words = "english")
    embeddings = vectorizer.fit_transform(text)
    embeddings = embeddings.toarray()
    return embeddings

# Average of Word2Vec
def getWord2Vec(text):
    embeddings = []
    for sentence in text:
        embeddings.append(Word2Vec_model(sentence).vector)
    return embeddings

# Universal Sentence Encoder
def getUSE(text):
    embeddings = []
    for sentence in text:
        embeddings.append(USE_model(sentence).vector)
    return embeddings
```

```python
# Clustering Models

def getClustersByThreshold(embeddings, linkageMethod, linkageMetric, threshold):
    clusters = linkage(embeddings, method = linkageMethod, metric = linkageMetric)
    fig = plt.figure(figsize=(25, 10))
    dn = dendrogram(clusters)
    plt.show()
    flatClusters = fcluster(clusters, t=threshold, criterion='distance')
    return flatClusters

def getClustersByNumOfClusters(embeddings, linkageMethod, linkageMetric, numOfCluster):
    clusters = linkage(embeddings, method = linkageMethod, metric = linkageMetric)
    fig = plt.figure(figsize=(25, 10))
    dn = dendrogram(clusters)
    plt.show()
    flatClusters = fcluster(clusters, t=numOfCluster, criterion='maxclust')
    return flatClusters
```

# Modified Silhouette Score

# Designed for Cosine Similarity

Paper: https://www.sciencedirect.com/science/article/pii/0377042787901257

Source Code: https://gist.github.com/AlexandreAbraham/5544803

```python
#Code for Modified Silhouette Score

def silhouette_score_block(X, labels, metric='euclidean', sample_size=None,
                           random_state=None, n_jobs=1, **kwds):
    if sample_size is not None:
        random_state = check_random_state(random_state)
        indices = random_state.permutation(X.shape[0])[:sample_size]
        if metric == "precomputed":
            raise ValueError('Distance matrix cannot be precomputed')
        else:
            X, labels = X[indices], labels[indices]
    return np.mean(silhouette_samples_block(
        X, labels, metric=metric, n_jobs=n_jobs, **kwds))

def silhouette_samples_block(X, labels, metric='euclidean', n_jobs=1, **kwds):
    A = _intra_cluster_similarity_block(X, labels, metric, n_jobs=n_jobs,
                                       **kwds)
    B = _farthest_cluster_similarity_block(X, labels, metric, n_jobs=n_jobs,
                                        **kwds)
    sil_samples = (A - B) / np.maximum(A, B)
    # nan values are for clusters of size 1, and should be 0
    return np.nan_to_num(sil_samples)

def _intra_cluster_similarity_block_(subX, metric, **kwds):
    similarity = cosine_similarity(subX)
    return similarity.sum(axis=1) / (similarity.shape[0] - 1)


def _intra_cluster_similarity_block(X, labels, metric, n_jobs=1, **kwds):
    intra_dist = np.zeros(labels.size, dtype=float)
    values = Parallel(n_jobs=n_jobs)(
            delayed(_intra_cluster_similarity_block_)
                (X[np.where(labels == label)[0]], metric, **kwds)
                for label in np.unique(labels))
    for label, values_ in zip(np.unique(labels), values):
        intra_dist[np.where(labels == label)[0]] = values_
    return intra_dist


def _farthest_cluster_similarity_block_(subX_a, subX_b, metric, **kwds):
    dist = cosine_similarity(subX_a, subX_b)
    dist_a = dist.mean(axis=1)
    dist_b = dist.mean(axis=0)
    return dist_a, dist_b


def _farthest_cluster_similarity_block(X, labels, metric, n_jobs=1, **kwds):
    inter_dist = np.empty(labels.size, dtype=float)
    inter_dist.fill(0)
    unique_labels = np.unique(labels)

    values = Parallel(n_jobs=n_jobs)(
            delayed(_farthest_cluster_similarity_block_)(
                X[np.where(labels == label_a)[0]],
                X[np.where(labels == label_b)[0]],
                metric, **kwds)
                for label_a, label_b in combinations(unique_labels, 2))

    for (label_a, label_b), (values_a, values_b) in \
            zip(combinations(unique_labels, 2), values):

            indices_a = np.where(labels == label_a)[0]
            inter_dist[indices_a] = np.maximum(values_a, inter_dist[indices_a])
            del indices_a
            indices_b = np.where(labels == label_b)[0]
            inter_dist[indices_b] = np.maximum(values_b, inter_dist[indices_b])
            del indices_b
    return inter_dist
```

```python
def evaluation(predictTags, tags, evalMetric):
    print(str(len(set(predictTags))) + " Clusters found.")

    result = {}
    result['Adjusted Rand Score'] = metrics.adjusted_rand_score(predictTags, tags)
    result['Homogenity'] = metrics.homogeneity_completeness_v_measure(predictTags, tags)[0]
    result['Completeness'] = metrics.homogeneity_completeness_v_measure(predictTags, tags)[1]
    result['V Measure'] = metrics.homogeneity_completeness_v_measure(predictTags, tags)[2]
    result['Silhouette Score'] = metrics.silhouette_score(embeddings, predictTags, metric=evalMetric)
    result['Silhouette Score_Amended'] = silhouette_score_block(embeddings, predictTags, metric=evalMetric)
    result_df = pd.DataFrame(data=result, index=[0])
    display(result_df)
    
    c_matrix = contingency_matrix(tags, predictTags)
    print("vertical length: " + str(len(c_matrix)))
    print("horizontal length: "+ str(len(c_matrix[0])))
    print(c_matrix)
    
    return c_matrix, result_df

def resultDic(predictTags, tags, news):
    dic = {}
    for i in range(len(predictTags)):
        if predictTags[i] in dic:
            dic[predictTags[i]][0].append(tags[i])
            dic[predictTags[i]][1].append(news[i])
        else:
            dic[predictTags[i]] = [[],[]]
            dic[predictTags[i]][0].append(tags[i])
            dic[predictTags[i]][1].append(news[i])
    return dic

def count(array):
    dic = {}
    for element in array:
        if element in dic:
            dic[element] += 1
        else:
            dic[element] = 1
    return dic
```

# Code Demonstration

```python
news, tags = getCSV()

# BERT
embeddings = getSentenceBERT(news)
bert_predictTags = getClustersByNumOfClusters(embeddings, "complete","cosine", 5)
embeddings = np.array(embeddings)
matrix_BERT, df_BERT = evaluation(bert_predictTags, tags, "cosine")

# TF_IDF
embeddings = getTFIDF(news)
tfidf_predictTags = getClustersByNumOfClusters(embeddings, "ward","euclidean", 5)
matrix_tfidf, df_tfidf = evaluation(tfidf_predictTags, tags, "euclidean")

# Word2Vec
embeddings = getWord2Vec(news)
w2v_predictTags = getClustersByNumOfClusters(embeddings, "complete","cosine", 5)
embeddings = np.array(embeddings)
matrix_W2V, df_W2V = evaluation(w2v_predictTags, tags, "cosine")

# USE
embeddings = getUSE(news)
use_predictTags = getClustersByNumOfClusters(embeddings, "complete","cosine", 5)
embeddings = np.array(embeddings)
matrix_USE, df_USE = evaluation(use_predictTags, tags, "cosine")

print("BERT:")
display(df_BERT)
print("TFIDF:")
display(df_tfidf)
print("Word2Vec:")
display(df_W2V)
print("USE:")
display(df_USE)

print("BERT:")
print(matrix_BERT)
print("TFIDF:")
print(matrix_tfidf)
print("Word2Vec:")
print(matrix_W2V)
print("USE:")
print(matrix_USE)
```
